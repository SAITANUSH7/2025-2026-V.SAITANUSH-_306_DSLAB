```python

# DATA CLEANING USING DATASET FROM URL

import pandas as pd
import numpy as np


url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)


print("Initial Shape:", df.shape)
print("\nMissing Values Before Cleaning:\n", df.isnull().sum())


df = df.dropna(subset=['Embarked'])


df['Age'] = df['Age'].fillna(df['Age'].median())


df = df.drop(columns=['Cabin'])


df = df.drop_duplicates()


df['Age'] = df['Age'].astype(int)
df['Fare'] = df['Fare'].astype(float)


df.rename(columns={
    'Pclass': 'Passenger_Class',
    'SibSp': 'Siblings_Spouses',
    'Parch': 'Parents_Children'
}, inplace=True)

df = df[df['Age'] >= 18]


df['Family_Size'] = df['Siblings_Spouses'] + df['Parents_Children']


df['Fare_Normalized'] = df['Fare'] / df['Fare'].max()


df['Age_Group'] = df['Age'].apply(lambda x: 'Young' if x < 30 else 'Adult')

# Final Output
print("\nFinal Shape After Cleaning:", df.shape)
print("\nMissing Values After Cleaning:\n", df.isnull().sum())
print("\nCleaned Dataset Preview:\n")
print(df.head())
```
