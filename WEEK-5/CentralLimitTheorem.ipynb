```python
#CENTRAL LIMIT THEOREM

import numpy as np
import math
x = np.array([10,12,16,44,55,60,70])
N = len(x)
print("Population:", x)
mu = np.sum(x) / N
print("Population mean (mu):", mu)
a2 = np.sum((x - mu)**2)
sigma2 = a2 / N
print("Population variance (sigma^2):", sigma2)
n = 3
var_xbar = sigma2 / n
print("Variance of sample mean:", var_xbar)
SE = math.sqrt(var_xbar)
print("Standard Error:", SE)
print("Distribution of x_bar ≈ N(", mu, ",", var_xbar, ")")
```
