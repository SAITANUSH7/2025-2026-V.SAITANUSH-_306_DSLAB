```python
#HYPOTHESIS.22

import numpy as np
import math
x = np.array([10,12,22,33,44,60,70])
N = len(x)
mu = np.sum(x)/N
a2 = np.sum((x - mu)**2)
sigma2 = a2/N
sigma = math.sqrt(sigma2)
sample = np.array([20,30,40])
n = len(sample)
x_bar = np.sum(sample)/n
Z = (x_bar - mu) / (sigma/math.sqrt(n))
print("Population mean (mu):", mu)
print("Population standard deviation (sigma):", sigma)
print("Sample mean (x_bar):", x_bar)
print("Z statistic:", Z)
critical_Z = 1.96
if abs(Z) <= critical_Z:
    print("Accept Null Hypothesis (H0)")
else:
    print("Reject Null Hypothesis (H0)")
```
