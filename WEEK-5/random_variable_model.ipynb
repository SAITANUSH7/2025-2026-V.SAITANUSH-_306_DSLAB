```python
#RANDOM VARIABLE MODEL

import numpy as np
import math
x = np.array([10,12,16,44,55,60,70])
n = len(x)
print("Random Variable:", x)
sample_mean = np.mean(x)
print("Sample Mean:", sample_mean)
EX = np.sum(x)/n
print("Expectation E(X):", EX)
a2 = np.sum((x - EX)**2)
variance = a2 / n
print("Variance:", variance)
```
