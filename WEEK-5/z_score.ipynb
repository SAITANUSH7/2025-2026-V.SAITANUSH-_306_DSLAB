```python
import numpy as np
data=[10,12,15,40,50]
mean=np.mean(data)
print("Mean:", mean)
std_dev=np.std(data,ddof=1)
print("Standard Deviation:", std_dev)
x=40
z_score=(x-mean)/std_dev
print("Z-Score of 40:",z_score)
```
