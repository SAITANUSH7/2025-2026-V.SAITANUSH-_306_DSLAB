```python
import numpy as np
import math
population=np.array([10,15,20,25,50,60,75])
N=len(population)
print("Population:",population)
mu=np.mean(population)
print("Population mean:",mu)
sigma2=np.sum((population-mu)**2)/N
sigma=math.sqrt(sigma2)
print("Population standard deviation:", sigma)
sample=np.array([20,30,40])
n=len(sample)
print("Sample:",sample)
x_bar=np.mean(sample)
print("Sample mean:",x_bar)
s2=np.sum((sample-x_bar)**2)/(n-1)
s=math.sqrt(s2)
print("Sample standard devition:",s)
SE=s/math.sqrt(n)
print("Standard Error:", SE)
Z=(x_bar-mu)/SE
print("Z score:", Z)
```
